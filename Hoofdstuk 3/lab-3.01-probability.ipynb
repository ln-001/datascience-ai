## Exercise 1.2 - Plot probability density

Make a graphical plot of the probability density function and the cumulative distribution function of a normal distribution with $\mu = 2.5$ and $\sigma = 1.5$.

What is the area under the density function between  x=0.5  and  x=4 . (Answer = 0.750)

```python

mu = 2.5 # Mean
sigma = 1.5 # Standard deviation
x = np. linspace(mu - 4 * sigma, mu + 4 * sigma, num=201)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y)
plt.title("Standard Normal Distribution")

z = stats.norm.cdf(x, mu, sigma)
plt.plot(x, z)


# Calculate the area
area = stats.norm.cdf(4, mu, sigma) - stats.norm.cdf(0.5, mu, sigma)
print("Area: " + str(area))
```

## Exercise 1.3 - Student t vs normal distribution
Make a plot of the probability density function and the cumulative distribution function for the Student's t-distribution with 3 degrees of freedom. 
Also plot the standard normal distribution, so you can compare both.

```python
from scipy.stats import norm, t

# Define the parameters for the distributions
df = 3  # degrees of freedom for the t-distribution

# Generate a range of values
x = np.linspace(-10, 10, 1000)

# Calculate the PDF and CDF for the standard normal distribution
pdf_normal = norm.pdf(x)
cdf_normal = norm.cdf(x)

# Calculate the PDF and CDF for the Student's t-distribution
pdf_t = t.pdf(x, df)
cdf_t = t.cdf(x, df)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot the PDFs
plt.subplot(1, 2, 1)
plt.plot(x, pdf_normal, label='Standard Normal PDF', color='blue')
plt.plot(x, pdf_t, label="Student's t (df=3) PDF", color='green')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

# Plot the CDFs
plt.subplot(1, 2, 2)
plt.plot(x, cdf_normal, label='Standard Normal CDF', color='blue')
plt.plot(x, cdf_t, label="Student's t (df=3) CDF", color='green')
plt.title('Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
